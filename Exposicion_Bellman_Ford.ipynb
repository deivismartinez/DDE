# **Implementación del Algoritmo Bellman-Ford en Python**

```python
class Graph:
    def __init__(self, vertices):
        self.M = vertices   # número total de vértices en el grafo
        self.graph = []     # Lista vacia para las aristas 
    # Agregar Aristas
    def add_edge(self, v_inicio, v_llegada, peso):
        self.graph.append([v_inicio, v_llegada, peso])

    # Imprimir la solución
    def print_solution(self, distance):
        sum = 0
        print("Vertice Distancia desde el origen")
        for k in range(self.M):
            print("{0}\t\t{1}".format(k, distance[k]))
            sum = sum + distance[k]
        print('Total = '+str(sum))

    def bellman_ford(self, origen):
        distance = [float("Inf")] * self.M
        distance[origen] = 0
        for _ in range(self.M - 1):
            for v_inicio, v_llegada, peso in self.graph:
                if distance[v_inicio] != float("Inf") and distance[v_inicio] + peso < distance[v_llegada]:
                    distance[v_llegada] = distance[v_inicio] + peso

        for v_inicio, v_llegada, peso in self.graph:
            if distance[v_inicio] != float("Inf") and distance[v_inicio] + peso < distance[v_llegada]:
                print("El grafo contiene ciclos de peso total negativo")
                return

        self.print_solution(distance)
```

```python
g = Graph(5)
g.add_edge(0, 1, 2)
g.add_edge(0, 2, 4)
g.add_edge(1, 3, 2)
g.add_edge(2, 4, 3)
g.add_edge(2, 3, 4)
g.add_edge(4, 3, -5)
g.bellman_ford(0)
```
